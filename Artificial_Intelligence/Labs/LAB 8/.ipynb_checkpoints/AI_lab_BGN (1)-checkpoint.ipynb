# Step 1: Import Libraries

```python

from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination
from pgmpy.estimators import BayesianEstimator
from pgmpy.sampling import BayesianModelSampling
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#Defining the structure (the "skeleton") of Bayesian Network.

```python
student_model = DiscreteBayesianNetwork([
    ('D', 'G'),   # Difficulty affects Grade
    ('I', 'G'),   # Intelligence affects Grade
    ('G', 'L'),   # Grade affects Letter of recommendation
    ('I', 'S')    # Intelligence affects SAT (Standardized Aptitude Test score)
])
```

 # Define the  Probabilities (CPDs)

```python

# P(D)
difficulty_cpd = TabularCPD(variable='D', variable_card=2,
                            values=[[0.6], [0.4]],
                            state_names={'D': ['Easy', 'Hard']})
# P(I)
intel_cpd = TabularCPD(variable='I', variable_card=2,
                       values=[[0.7], [0.3]],
                       state_names={'I': ['Low', 'High']})

# P(G | I, D)
grade_cpd = TabularCPD(variable='G', variable_card=3,
                       values=[[0.7, 0.9, 0.1, 0.2],  # G=0 (Low)
                               [0.2, 0.08, 0.3, 0.3], # G=1 (Medium)
                               [0.1, 0.02, 0.6, 0.5]],# G=2 (High)
                       evidence=['I', 'D'],
                       evidence_card=[2, 2],
                       state_names={'G': ['G_Low', 'G_Medium', 'G_High'],
                                    'I': ['Low', 'High'],
                                    'D': ['Easy', 'Hard']})

# P(L | G)
letter_cpd = TabularCPD(variable='L', variable_card=2,
                        values=[[0.9,0.3, 0.1],  # L=0 (Weak)
                                [0.1,0.7, 0.9]], # L=1 (Strong)
                        evidence=['G'],
                        evidence_card=[3],
                        state_names={'L': ['L_Weak', 'L_Strong'],
                                     'G': ['G_Low', 'G_Medium', 'G_High']})

# P(S | I)
sat_cpd = TabularCPD(variable='S', variable_card=2,
                     values=[[0.95, 0.2],  # S=0 (Weak)
                             [0.05, 0.8]],  # S=1 (Strong)
                     evidence=['I'],
                     evidence_card=[2],
                     state_names={'S': ['S_Weak', 'S_Strong'],
                                  'I': ['Low', 'High']})
```

Rows: Represent the state of the 'Grade' (G).

Row 1: G = G_Low

Row 2: G = G_Medium

Row 3: G = G_High

Columns: Represent every possible combination of the "evidence" variables 'I' (Intelligence) and 'D' (Difficulty). There are 4 columns with the possible order:

(I='Low', D='Easy')

(I='High', D='Easy')

(I='Low', D='Hard')

(I='High', D='Hard')

Examples:

0.7 (Row 1, Column 1): This is the probability of getting a 'G_Low' grade given that Intelligence is 'Low' AND Difficulty is 'medium'. P(G='G_Low' | I='Low', D='Easy') = 0.7 (or 70%)

0.6 (Row 3, Column 3): This is the probability of getting a 'G_High' grade given that Intelligence is 'Low' AND Difficulty is 'Hard'. P(G='G_High' | I='Low', D='Hard') = 0.6 (or 60%)

0.2 (Row 1, Column 4): This is the probability of getting a 'G_LOW' grade given that Intelligence is 'High' AND Difficulty is 'Easy'. P(G='G_High' | I='High', D='Easy') = 0.2 (or 20%)

# Add CPDs to the Model


```python

student_model.add_cpds(difficulty_cpd, intel_cpd, grade_cpd, letter_cpd, sat_cpd)
```

#Validate the Model

```python


print("Model check:", student_model.check_model())
```

#Create the Directed Graph (DiGraph) Using the edges from student_model.

```python

graph = nx.DiGraph()
graph.add_edges_from([
    ('I', 'G'),  # Intelligence -> Grade
    ('D', 'G'),  # Difficulty -> Grade
    ('G', 'L'),  # Grade -> Letter
    ('I', 'S')   # Intelligence -> SAT
])

# Define positions for a  diagram
pos = {
    'I': (0, 1),    # Intelligence
    'D': (2, 1),    # Difficulty
    'G': (1, 0.5),  # Grade
    'L': (1, -0.2), # Letter
    'S': (-1, 0.5)  # SAT
}

#  Draw the directed graph
plt.figure(figsize=(6, 4))
nx.draw_networkx_nodes(graph, pos, node_color='lightblue', node_size=1600, edgecolors='black')
nx.draw_networkx_labels(graph, pos, font_size=12, font_weight='bold')

# Add 'node_size=1600' here to make the arrows visible.
nx.draw_networkx_edges(
    graph,
    pos,
    arrows=True,
    arrowstyle='-|>',
    arrowsize=20,
    width=2,
    node_size=1600
)
plt.title("Student Bayesian Network", fontsize=14)
plt.axis('off')
plt.tight_layout()
plt.show()










```

# Perform Inference

```python

infer = VariableElimination(student_model)
```

```python
query_1 = infer.query(variables=['L'], evidence={'I': 'High'})
print(query_1)
```

```python
query_1 = infer.query(variables=['L'], evidence={'I': 'Low'})
print(query_1)
```

```python
query_2 = infer.query(variables=['I'], evidence={'L': 'L_Strong'})
print(query_2)
```

```python
query_3 = infer.query(variables=['G'], evidence={'I': 'High', 'D': 'Easy'})
print(query_3)
```


#Additional Task:
#  Generate Synthetic Data from the Model
# We  collect 10,000 data points (e.g., from 10,000 students)
# that follow the probabilities we defined above.


```python
print("\n Generating 10,000 Data Samples ---")

#  Set up the sampler
sampler = BayesianModelSampling(student_model)

#  Generate the data
data = sampler.forward_sample(size=10000)

print("Data generated. Here are the first 5 rows:")
print(data.head())
```

#  Learn Parameters from Data
# Now, we pretend we are researchers who were *given* the synthetic data.
# We know the graph structure (the arrows), but not the probabilities.
# We will use .fit() to "learn" the probabilities from the data.


```python
print("\n Learning Parameters from Data ---")

#  Create a new, "empty" model with only the structure
# This model doesn't know any of the probabilities yet.


model_to_learn = DiscreteBayesianNetwork(student_model.edges())

# "Train" the model on the data.
# The .fit() method calculates all the conditional probabilities from the data.

model_to_learn.fit(data, estimator=BayesianEstimator)

print("Model has been 'fit' to the data.")
```


#  Compare the "original" vs. "Learned" CPDs

```python
print("\n--- PART 4: Comparing original vs. Learned")

#  Compare P(I)
print("\nIntelligence CPD ")
print("Original P(I=High):")
print(student_model.get_cpds('I').values[1]) # This was correct
print("Learned P(I=High):")
print(model_to_learn.get_cpds('I').values[1]) # This was correct
```

```python
#  Compare P(S | I)
print("\n### SAT CPD ###")
print("Original P(S=Strong | I=High):")
print(student_model.get_cpds('S').values[1][1])
print("Learned P(S=Strong | I=High):")
print(model_to_learn.get_cpds('S').values[1][1])
```

```python
#  Compare a complex one: P(G | I, D)
print("\n### Grade CPD ###")
print("Original P(G=High | I=High, D=Less):")
print(student_model.get_cpds('G').values[2][1][0]) # (G=0, I=1, D=0)
print("Learned P(G=High | I=High, D=Less):")
print(model_to_learn.get_cpds('G').values[2][1][0]) # (G=2, I=1, D=0)
```

# Perform Inference from learned model

```python
infer_learned_model = VariableElimination(model_to_learn)
```

```python
query_1_from_learned_model = infer_learned_model.query(variables=['L'], evidence={'I': 'Low'})
print(query_1_from_learned_model)
```

```python
query_2_from_learned_model = infer_learned_model.query(variables=['L'], evidence={'I': 'High'})
print(query_2_from_learned_model)
```
