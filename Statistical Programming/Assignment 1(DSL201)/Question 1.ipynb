```python
import pandas as pd
import numpy as np
with open("fileA.txt", 'r', encoding='utf-8') as file:
    
    
        text = file.read()
        text = text.replace('\ufeff', "").lower()
        text = text.replace(',', "").lower()    

l=[]
for i in text:
    l.append(i)
a=[0]*26
for i in l:
    a[ord(i)-ord('a')]=a[ord(i)-ord('a')]+1
li=[]     
for i in range(0,26):
    li.append((chr(i+ord('a')),a[i]))
df=pd.DataFrame(li,a,columns=['Alphabets','Frequency'])
ll=len(l)
#print(ll)
print("The probablities of the wrods are as follows : ")
df['Frequency']=df['Frequency']/ll
df=df.reset_index(drop=True)
df.index=df.index+1
print(df)
df.sort_values('Frequency',ascending=False,inplace=True)
df=df.reset_index(drop=True)
df.index=df.index+1
print("TOP 10 Alphabets are :")
df['Alphabets'].head(10)
```

```python
import pandas as pd
import math
with open("fileB.txt", 'r', encoding='utf-8') as file:
    
    
        text = file.read()
l=[]
for i in text:
    if (ord(i)-ord('a')>=0 and ord(i)-ord('a')<=26 and ord(i)-ord('a')==int(ord(i)-ord('a'))):
        l.append(i)
a=[0]*26
for i in l:
    if 'a' <= i <= 'z':
        a[ord(i) - ord('a')] += 1    
li=[]     
for i in range(0,26):
    li.append((chr(i+ord('a')),a[i]))
df=pd.DataFrame(li,a,columns=['Alphabets','Frequency'])
ll=len(l)
#print(ll)
print("The probablities of the words are as follows : ")
df['Probablity']=df['Frequency']/ll
df=df.reset_index(drop=True)
df.index=df.index+1
print(df)
h=0
for i in df['Probablity']:
    if i!=0:
        h=h-i*(math.log(i))        
print(f"The entropy is given by: {h} ")        
```

```python
import pandas as pd
import math
with open("fileC.txt", 'r', encoding='utf-8') as file:
    text = file.read().lower()
    text=text.split()
l=[]
for i in text:
    l.append(i)
d={}
for i in l:
    if i not in d.keys():
        d[i]=1
    else:
        d[i]=d[i]+1
ll=len(l)
li=list(d.keys())
a=list(d.values())
df=pd.DataFrame(li,a,columns=['Words'])
df['Frequency']=df.index
df=df.reset_index(drop=True)

df.index=df.index+1
df['Probablity']=df['Frequency']/ll
print("The probablities of the words are as follows : ")
print(df)
df.sort_values('Frequency',ascending=False,inplace=True)
#print(df)
print("TOP 10 Words are :")
df['Words'].head(10)
```
