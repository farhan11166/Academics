```python
import random
import statistics as stat
import seaborn as sns
import pandas as pd
import math
l=[]
for i in range(5,1000,5):
  a=[]
  for j in range(0,i):

    a.append(random.gauss(4,3))
  l.append(a)
f=[]
r=[]
h=[]
for i in l:

  f.append(len(i))
  r.append(stat.mean(i))
  h.append(stat.stdev(i))

df=pd.DataFrame({'no. of points':f,'mean':r,'standard deviation':h})
print(sns.scatterplot(x='no. of points',y='mean',data=df))
import matplotlib.pyplot as plt
plt.figure()
print(sns.scatterplot(x='no. of points',y='standard deviation',data=df))
```
