```python
import random
import statistics as stat
import seaborn as sns
import pandas as pd
l=[]
for i in range(5,1000,5):
  a=[]
  for j in range(0,i):

    a.append(random.uniform(0,1))
  l.append(a)
#print(l)
f=[]
r=[]
h=[]
for i in l:

  f.append(len(i))
  r.append(stat.mean(i))
  h.append(stat.stdev(i))

df=pd.DataFrame({'no. of points':f,'mean':r,'standard deviation':h})
m1=sns.scatterplot(x='no. of points',y='mean',data=df)
print(m1)
import matplotlib.pyplot as plt
plt.figure()
print(sns.scatterplot(x='no. of points',y='standard deviation',data=df))
```
