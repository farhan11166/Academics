```python
import math
from math import comb
import seaborn as sns
import pandas as pd
l=[]
for i in range(0,16):
  l.append([i,comb(15,i)*pow(0.2,i)*pow(0.8,15-i)])
df=pd.DataFrame(l,columns=['no.of success','probablity'])
sns.barplot(df,x='no.of success',y='probablity')
```

```python

j=0.1
import matplotlib.pyplot as plt
for k in range(0,8):
  l=[]

  for i in range(0,16):
    l.append([i,comb(15,i)*pow(j,i)*pow(1-j,15-i)])
  #s=comb(15,i)*pow(0.2,i)*pow(0.8,15-i)
   #print(s)
  df=pd.DataFrame(l,columns=['no.of success','probability'])
  plt.figure()
  sns.barplot(data=df, x='no.of success', y='probability')
  plt.title(f"Binomial Distribution (p = {j:.1f})")
  plt.xlabel("No. of Successes")
  plt.ylabel("Probability")
  plt.figure()

  j=j+0.2
  if(j>0.9):
    break
```
