```python
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt   
import pandas as pd
import numpy as np
import seaborn as sns
scaler = MinMaxScaler(feature_range=(-1, 1))
def normalizedataframe(df):
    return (2*df1-(df1.max()+df1.min()))/(df1.max()-df1.min())

df1=pd.read_csv('dataset_q1/data_1 - data_1.csv')
df2=pd.read_csv('dataset_q1/data_2 - data_2.csv')
l=df1.columns
g=df2.columns
l1=[]
for i in l:
    if i=='1398.697391.1':
        l1.append(1398.697391)
        
    else:
        l1.append(float(i))
l2=[]
for i in g:
    if i=='1850.697391.1':
        l2.append(1850.697391)
        
    else:
        l2.append(float(i))
ll=[]
ll.append(l1)
ll2=[]
ll2.append(l1)
df1= df1.values.tolist()
df2= df2.values.tolist()
for i in df1:
    ll.append(i)    
for j in df2:
    ll2.append(j)
ll=np.array(ll)
ll2=np.array(ll2)
```

```python
corr_matrix = np.corrcoef(ll, rowvar=True)
corr_matrix=pd.DataFrame(corr_matrix)
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False)
```
