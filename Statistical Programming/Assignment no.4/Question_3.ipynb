```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as st
df=pd.read_csv('dataset_q3/gaussian_dataset_with_noise.csv')
df1=pd.read_csv('dataset_q3/gaussian_dataset_with_noise.csv')
df=df.values.tolist()
df=np.array(df)
mean=np.mean(df)
print(f"the mean is {mean}")
var=np.var(df)
std=(var)**(1/2)
print(f"the standard deviation is {std}")
for i in range(1,4):
    k=(std*i)
    l=[i for i in df if abs(mean-i)<=k]
    print(f"the percent of element in {i}sigma distance from mean is ={len(l)/10}")
import seaborn as sns
sns.kdeplot(df)
df=df.reshape(-1)
from statsmodels.distributions.empirical_distribution import ECDF
df=np.sort(df)
ecdf = ECDF(df)
x = np.linspace(min(df), max(df), 100)
y = ecdf(x)
plt.figure()
plt.plot(x, y, marker='.', linestyle='none')
plt.title("Empirical CDF (ECDF)")
plt.xlabel("Value")
plt.ylabel("CDF")
plt.grid(True)
plt.show()
p1=0
p2=0
for i in df:
    if i>mean+2*std:
        break
    if i<=mean-2*std:
        p1=ecdf(i)
    if i>mean-2*std and i<=mean+2*std:
        p2=ecdf(i)
print(f"the probablity of data points lying beyond 2sigma ={1-(p2)}")
```
