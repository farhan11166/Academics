```python
import numpy as np
import matplotlib.pyplot as plt
p=[10,100,1000,10000]
for i in p:


  k=np.random.exponential(scale=1/2,size=i)
  l=(1-np.exp(-2*k))
  m=np.random.uniform(0,1,size=i)
  plt.figure(figsize=(12, 6))
  plt.subplot(1,2,1)
  plt.hist(l,bins=20,density=True,color='orange')
  plt.title(f"exponential Process Histogram (n={i})")
  plt.xlabel("Y")
  plt.ylabel("distibution")
  #plt.figure()
  plt.subplot(1,2,2)
  plt.hist(m,bins=20,density=True,color='red')
  plt.title(f"Uniform Process Histogram (n={i})")
  plt.xlabel("Y")
  plt.ylabel("distibution")
  plt.show()






```

```python
import re
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
with open('text_file.txt','r') as file:
  text=file.read()
  words=re.findall(r'\b\w+\b', text.lower())
  counts=Counter(words)
l=pd.DataFrame(counts.items(),columns=['words','frequency'])
l=l.sort_values('frequency',ascending=False)
f=l.head(30)
plt.figure(figsize=(30,12))
plt.subplot(1,2,1)
plt.bar(f['words'],f['frequency'],color='blue')
plt.title("Original Word freq")
plt.xticks(rotation=90)
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.subplot(1,2,2)
k=0
q=[]
for i in f['frequency']:
  k=k+i
  q.append(k)
for i in q:
  i=i/k

plt.bar(f['words'],q,color='blue')
plt.title("CDF-Transformed Word freq")
plt.xlabel("Words")
plt.ylabel("Transformed Frequency (CDF)")

```

```python
import numpy as np
import matplotlib.pyplot as plt


n=100000
l_exp=1


U=np.random.uniform(0,1,n)
Y=-np.log(1-U)/l_exp


X=np.sqrt(U)

plt.subplot(1,2,1)
plt.hist(Y,bins=30,density=True,color='skyblue',edgecolor='black')
plt.title(" Exponential Distribution log(1-x)")
plt.xlabel("Y")
plt.ylabel("Density")


plt.subplot(1,2,2)
plt.hist(X,bins=30,density=True,color='orange',edgecolor='black')
plt.title("Triangular Distribution sqrt(x)")
plt.xlabel("Y")
plt.ylabel("Density")

plt.tight_layout()
plt.show()

```
