```python
from math import inf
state = (3, 4, 5)   


def successors(state):
    states = []
    for i, heap in enumerate(state):
        for remove in range(1, heap + 1):
            new_heap = list(state)
            new_heap[i] -= remove
            states.append(tuple(sorted(new_heap)))  
    return states
def is_terminal(state):
    return all(h == 0 for h in state)
def utility(state, maximizing_player):
    if is_terminal(state):
        return -1 if maximizing_player else +1
    return None
explored_nodes = 0
pruned_nodes = 0

def alphabeta(state, alpha, beta, maximizing):
    global explored_nodes, pruned_nodes
    explored_nodes += 1

    util = utility(state, maximizing)
    if util is not None:
        return util, None

    if maximizing:
        value, best_move = -inf, None
        for move in successors(state):
            print(f"Considering move: {state} → {move}")
            child_value, _ = alphabeta(move, alpha, beta, False)
            if child_value > value:
                value, best_move = child_value, move
            alpha = max(alpha, value)
            if alpha >= beta:
                print(f"[Pruned branch at {move} because alpha >= beta]")
                pruned_nodes += 1
                break
        return value, best_move
    else:
        value, best_move = inf, None
        for move in successors(state):
            print(f"Considering move: {state} → {move}")
            child_value, _ = alphabeta(move, alpha, beta, True)
            if child_value < value:
                value, best_move = child_value, move
            beta = min(beta, value)
            if alpha >= beta:
                print(f"[Pruned branch at {move} because alpha >= beta]")
                pruned_nodes += 1
                break
        return value, best_move
print(f"Initial State: {state}\n")

value, best = alphabeta(state, -inf, inf, True)

print(f"\nBest Move for MAX: {state} → {best}")
print("Outcome:", "Winning position" if value == 1 else "Losing position")
print(f"Nodes Explored: {explored_nodes}")
print(f"Nodes Pruned: {pruned_nodes}")
```
